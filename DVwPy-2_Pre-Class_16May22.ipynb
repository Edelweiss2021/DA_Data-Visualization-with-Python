### **Matplotlib-Styling**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
%matplotlib inline
```

```python
style.use('ggplot')
x = np.arange(0,30,0.2)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
style.use('fivethirtyeight')
x = np.arange(0,30,0.2)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
style.use('dark_background')
x = np.arange(0,30,0.2)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
style.use('seaborn')
x = np.arange(0,30,0.2)
y1 = np.sin(x)
y2 = np.cos(x)

#plt.grid(True)

plt.title("Sine Function")

# plt.suptitle("I am bigger title") #super title
# plt.xlabel("Height of Students")
# plt.ylabel("Weight of Students")
plt.legend(loc="upper left")

plt.plot(x,y1, label="Sine")
plt.plot(x,y2, label="Cosine")

plt.show()
```

```python
x = np.linspace(0, 5, 11)
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, x+13, color="green", lw=3, ls="-", marker = "o", markersize=10)
ax.set_xlim([0,3])
ax.set_ylim([13, 16])
```

```python
x = np.linspace(0,5,11)
fig, axes = plt.subplots()
axes.plot(x, x**2, label="square")
axes.plot(x, x**3, label="cube")
axes.legend()
plt.savefig("square_cube.png")
```

### **Matplotlib Plotting Type**

**Bar Chart**

matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, align='center', data=None)

```python
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
profit = [40, 2, 34, 12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos, company)
plt.xlabel("revenue($bn)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2, revenue, width=0.4, label="Revenue")

plt.bar(xpos+0.2, profit, width=0.4, label="Profit")
plt.legend();
```

```python
plt.yticks(xpos, company)
plt.xlabel("revenue($bn)")
plt.title("US Tech Stocks")
plt.barh(xpos-0.2, revenue, label="Revenue")

plt.barh(xpos+0.2, profit, label="Profit")
plt.legend();
```
