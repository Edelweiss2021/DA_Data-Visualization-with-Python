### **Seaborn - Comparison Plots**

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import style
import matplotlib

import seaborn as sns

%matplotlib inline
```

#### **catplot**

`seaborn.catplot` (x=None, y=None, hue=None, data=None, row=None, col=None, 
estimator=function, palette=None, legend=True)

```python
#load car data and drop rows with missings
cars = sns.load_dataset('mpg').dropna()

#filter to 4-, 6-, 8-cylinder cars
cars = cars[cars.cylinders.isin([4, 6, 8])]

#add old/new column
cars["type"] = ['old' if x<=76 else "new" for x in cars.model_year]
```

```python
cars.head()
```

```python
cars.shape
```

#### **Basics**

```python
sns.set_style('white')
```

```python
sns.catplot(y="horsepower", data=cars); # vertical stripplot, use 'x' for horizontal
```

```python
sns.catplot(x="horsepower", data=cars);
```

##### **Kind**

```python
sns.catplot(x="horsepower", data=cars, kind='box');
```

```python
sns.catplot(x="horsepower", data=cars, kind='violin');
```

##### **Categorical Variables, Hue**

```python
sns.catplot(x="origin", y='horsepower',data=cars, kind='box');
```

#### **jointplot**

`seaborn.jointplot` (x=None, y=None, data=None, kind='scatter', hue=None)

```python
geyser = sns.load_dataset('geyser')
```

```python
geyser.head()
```

```python
geyser.shape
```

#### **Basics**

```python
sns.set_style('darkgrid')
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser);
```

#### **Plot Kinds**

There are currently six kinds of jointplots allowed in seaborn:

1. scatter (default)
2. kde
3. reg
4. hist
5. hex
6. resid

##### **scatter (default)**

```python
sns.jointplot(x='waiting', y='duration', data=geyser, kind='scatter');
```

##### **kde**

```python
sns.jointplot(x='waiting', y='duration', data=geyser, kind='kde');
```

##### **reg**

```python
sns.jointplot(x='waiting', y='duration', data=geyser, kind='reg');
```

##### **hist**

```python
sns.jointplot(x='waiting', y='duration', data=geyser, kind='hist');
```

##### **Returns JointGrid**

```python
g = sns.jointplot(x='waiting', y='duration', data=geyser);
```

```python
type(g)
```

```python
g.plot_joint(sns.kdeplot, color='gray', levels=5)
plt.show();
```

##### **hue**

#### **Styling**

##### **color, palette**

```python
sns.jointplot(x='waiting', y='duration', data=geyser, color='purple');
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser, hue='kind', palette='pastel');
```

##### **height, space, ratio**

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              height=8,   #Default: 6
              space=0.1,  #Default: 0.2
              ratio=2     #Default: 5
             );
```

##### **joint_kws**

Joint keywords should match the figure type in the middle of the jointplot.

```python
sns.jointplot(x='waiting', y='duration', data=geyser, joint_kws={'marker': 5});
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              kind='reg',
              joint_kws={'ci': None, 'order': 2}
             );
```

##### **marginal_kws**

Marginal keywords should match the type of plot for the marginal distribution plots (the outside plots).

```python
sns.jointplot(x='waiting', y='duration', data=geyser, marginal_kws={'color':'xkcd:golden'});
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser, kind='kde', marginal_kws={'lw':4, 'shade':True});
```
