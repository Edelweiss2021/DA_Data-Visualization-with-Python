Bar plots

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head(7)
```

# 2. Make a bar plot

```python
df.plot.bar();
```

```python
df.plot(kind='bar');
```

# 3. Stack the bars

```python
df.plot.bar(stacked=True);
```

# 3. Rotate the labels and set figsize

```python
df.plot.bar(rot=0, stacked=True, figsize = (9,6));
```

# 4. Replace the labels by "Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"

```python
# primary solution
labels = ["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"]
df["labels"] = labels
df
```

```python
df.set_index('labels', inplace=True)
df
```

```python
# alternate solution
df.index = ["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"]
df
```

# 5. Unstuck the bars, annotate the hight of the bars on top of them

```python
labels=["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"]

g = df.plot.bar(figsize=(9,6), rot=0)
g.set_xticklabels(labels)
for p in g.patches:
    g.annotate((p.get_height()), (p.get_x()+0.02, p.get_height()+0.5))
```

# 7. Plot the avg tip by size  (lineplot)

```python
tips.groupby("size")["tip"].mean().plot(); #pandas
```

```python
tips.columns
```

```python
sns.lineplot(data=tips, x='size', y="tip", ci=None); #seaborn
```

```python
df1= tips.groupby('size')['tip'].mean() # matplotlib

fig, ax = plt.subplots()
ax.plot(df1.index, df1.values)
ax.set_xlabel("size");
```
