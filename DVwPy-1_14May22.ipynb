### **Matplotlib-Figure and Axes**

#### **Format strings in plot function**

```python
plt.plot(x,y, 'm+-.') #format string method
```

```python
plt.plot(x,y, color='magenta', marker='+', linestyle='dashdot' )   #keyword argument
```

```python
plt.plot(x,y, color='#008040', marker='+', linestyle='dashdot', markersize=10 )   #keyword argument, color as hexadecimal
```

```python
plt.plot(x,y, color='#008040', alpha=0.3) # alpha: transparency
```

#### **Axes labels, Legend, Grid**

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")

plt.legend(loc="best", shadow=True, fontsize='small') # loc default = "best"

plt.grid() # adding gridlines
```

### **Subplots with Matplotlib**

**Functional Method**

plt.subplot(2,1,1)

plt.plot(x1, y1)

plt.subplot(2,1,2)

plt.plot(x2, x2)

```python
x = np.linspace(0, 10, 10)
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sqrt(x)

plt.figure()
plt.subplot(2,2,1)
plt.plot(x, y1, 'ro')
plt.title("$y_1=x$")


plt.subplot(2,2,2)
plt.plot(x, y2, 'g--')
plt.title("$y_2=x^2$")

plt.subplot(2,2,3)
plt.plot(x, y3, 'b^')
plt.title("$y_3=x^3$")

plt.subplot(2,2,4)
plt.plot(x, y4, 'ks')
plt.title("$y_4=\sqrt{x}$")
```

```python
plt.figure()
plt.subplot(4,1,1)
plt.plot(x, y1, 'ro')
plt.title("$y_1=x$")


plt.subplot(4,1,2)
plt.plot(x, y2, 'g--')
plt.title("$y_2=x^2$")

plt.subplot(4,1,3)
plt.plot(x, y3, 'b^')
plt.title("$y_3=x^3$")

plt.subplot(4,1,4)
plt.plot(x, y4, 'ks')
plt.title("$y_4=\sqrt{x}$")
```
